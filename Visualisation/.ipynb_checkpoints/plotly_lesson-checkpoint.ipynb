# Часть 1. Обработка данных
Импортируем данные и смотрим их

```python
data = pd.read_csv("public.adidas_df.csv")
data.head()
```

```python
data.info()
```

Проведем ряд преобразований для генерации новых фичей

```python
data["sell_date"] = pd.to_datetime(data["sell_date"])
data["operating_margin_share"] = (
    data["operating_margin_share"].str.replace(",", ".").astype(float)
)
data = data.sort_values(by="sell_date")
data.head()
```

```python
data["sell_month"] = data["sell_date"].dt.month_name()
data["sell_month_num"] = data["sell_date"].dt.month
data["sell_month"] = data["sell_month"].apply(lambda x: x[:3])
data["sell_year"] = data["sell_date"].dt.year.astype("str")
data["sell_year_month"] = data["sell_year"] + "-" + data["sell_month"]
data["sell_day_of_week"] = data["sell_date"].dt.day_of_week
data["revenue"] = data["price_per_unit"] * data["units_sold"]
data["operating_margin"] = data["operating_margin_share"] * data["revenue"]

data.head()
```

# Часть 2. Создание графиков

### Динамика выручки и маржи


```python
data_year_month = data.pivot_table(
    values=("revenue", "operating_margin"),
    index=["sell_year", "sell_month_num", "sell_year_month"],
).reset_index()
data_year_month
data_year_month["operating_margin_share"] = (
    data_year_month["operating_margin"] / data_year_month["revenue"]
)
data_year_month.head()
```

### Подневная динамика выручки и маржи

```python
data_date = pd.pivot_table(
    data=data, values=("operating_margin", "revenue"), index="sell_date"
).reset_index()
data_date["operating_margin_share"] = (
    data_date["operating_margin"] / data_date["revenue"]
)
data_date["revenue_rolling"] = data_date["revenue"].rolling(7).mean()
data_date["operating_margin_rolling"] = data_date["operating_margin"].rolling(7).mean()
data_date["operating_margin_share_rolling"] = (
    data_date["operating_margin_rolling"] / data_date["revenue_rolling"]
)

data_date.head(10)
```
