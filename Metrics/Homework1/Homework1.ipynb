```python
import pandas as pd
import numpy as np
```

```python
dfo = pd.read_csv("orders.csv")
```

```python
dfo.head()
```

```python
dfc = pd.read_csv("customers.csv")
cols = ['customer_id']
cols.extend(dfc.columns[1:])
dfc.columns = cols
```

```python
dfc.head()
```

```python
dfo['order_date']
```

```python
dfo['order_month'] = dfo['order_date'].apply(lambda x: x[5:7])
dfo['order_year'] = dfo['order_date'].apply(lambda x: x[0:4])
```

```python
dfo.head()
```

```python
df_merge = pd.merge(dfo, dfc, 'inner', on='customer_id')
```

```python
df_merge.sample(5)
```

```python
df_first = df_merge.groupby('customer_id')['order_year'].min().reset_index()
df_first.columns = ['customer_id', 'first_order']
```

```python
df_first.sample(5)
```

```python
df_full = pd.merge(df_merge, df_first, 'inner', on='customer_id')
df_full.sample(5)
```

```python
df_users = df_full.groupby(['customer_id', 'first_order', 'order_year'])\
    .agg({'sales':'sum', 'id':'count'}).reset_index()
df_users.sample(5)
```
