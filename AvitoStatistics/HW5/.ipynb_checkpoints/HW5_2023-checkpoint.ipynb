## Задача 1 (4 балла)

Интересно, насколько применение поправки Бонферрони увеличивает ширину доверительного интервала?

Допустим, что в нашем эксперименте мы строим двусторонний 95% доверительный интервал для $\mu$ с помощью t-test (размер выборки 1 миллион).

Пусть m - количество гипотез.

Для разных вариантов m продемонстрировать, насколько расширяется доверительный интервал для $\mu$ в этом эксперименте.

Продемонстрируйте зависимость на графиках, а также не забудьте написать вывод.

```python
from scipy import stats as sps
import numpy as np
import matplotlib.pyplot as plt
```

Стоит сказать, что размер выборки у нас миллион, поэтому нет смысла смотреть на разные распределения из-за ЦПТ, поэтому я рассматриваю нормальное сразу.

```python
SIZE = 1000000

data = sps.norm.rvs(loc=45, scale=500, size=SIZE, random_state=80)
print(sps.t.interval(alpha=0.95, loc=np.mean(data), df=SIZE-1, scale=sps.sem(data)))
print(sps.t.interval(alpha=0.975, loc=np.mean(data), df=SIZE-1, scale=sps.sem(data)))


# Это я сделал, чтобы сравнить ДИ по определению и ДИ, рассчитанный с помощью t.interval
mu_left, mu_right = 0, 0
for mu in np.linspace(40, 50, 1000):
    res = sps.ttest_1samp(data, mu)
    if mu_left == 0 and res.pvalue > 0.05:
        mu_left = mu
    if res.pvalue < 0.05 and mu_left != 0:
        break
    mu_right = mu
print((mu_left, mu_right))
```

Сформируем список разниц в ДИ в зависимости от количества гипотез

```python
list_m = np.arange(1, 100)
l_b, r_b = sps.t.interval(alpha=0.95, loc=np.mean(data), df=SIZE-1, scale=sps.sem(data))
length_ci = r_b - l_b

diff_list = []
for m in list_m:
    l_b_m, r_b_m = sps.t.interval(alpha=1 - 0.05/m, loc=np.mean(data), df=SIZE-1, scale=sps.sem(data))
    diff_list.append((r_b_m - l_b_m) - length_ci)
```

```python
plt.figure(figsize=(10, 5))
plt.title('Расширение CI от увеличения гипотез')
plt.plot(list_m, diff_list, label='norm')
plt.xlabel('Количество гипотез m')
plt.ylabel('Расширение CI')
plt.legend(loc='best')
```

Видим, что увеличение подчиняется логарифмическому закону. Это происходит из-за того, что мы сначала "шагаем" по alpha на 2.5, потом на 1.66, и т.д. , постепенно уменьшая шаг, тем самым уменньшая охваченные значения у t распределения.

### Пункт А. 2 балла

В данном домашнем задании вам предлагается придумать относительный t-test критерий для AB-тестов.


Какую гипотезу мы проверяли ранее в t-test?

$H_0: E T - E C = 0$

Мы умееем для такой гипотезы строить критерий и доверительный интервал.

Но в [бизнесе чаще бывает полезна](https://habr.com/ru/companies/avito/articles/571094/) другая формулировка:

$H_0: \dfrac{E T}{E C} - 1= 0$

С точки зрения гипотезы она ничем не отличается от перового случая. Но вот доверительный интервал будет другим! И нам интересно научиться строить корректный доверительный интервал для этой гипотезы.


В данном пункте предлагается построить этот критерий в случае, если выборки тест и контроля одинакового размера.

P.S. Решить задачу нужно именно через линеаризацию




```python
import scipy.stats as sps
import numpy as np
from tqdm.auto import tqdm
from statsmodels.stats.proportion import proportion_confint
# ваши импорты
#import warnings
#warnings.filterwarnings("ignore")
```

Распишите текстом ваш алгоритм действий.

Из случайной величины T сэмплируются $T_1$, ..$T_n$
Из случайной величины C сэмплируются $C_1$, ..$C_n$

Оценка $\frac{T_1+...+T_n}{C_1+...+C_n}$ будет несмещенной оценкой $\frac{ET}{EC}$

Обозначим за $X = T_1+...+T_n$, за $Y = C_1+...+C_n$

В моей терминалогии переменная с крышечкой - это уже реализованная случайная величина.

Итак, наша $f(X, Y) = \frac{X}{Y}$, разложим ее в окрестности точки $(\hat X, \hat Y)$, а так как $(X - \hat X)$ и $(Y - \hat Y)$ близки к нулю по ЗБЧ, то мы используем основные формулы Маклорена (Тейлор в нуле).

$$f(X, Y) = \frac{\hat X}{\hat Y} + (X - \hat X)\frac{\partial f}{\partial X}(\hat X, \hat Y) + (Y - \hat Y)\frac{\partial f}{\partial Y}(\hat X, \hat Y) = \frac{\hat X}{\hat Y} + (X - \hat X)\frac{1}{\hat Y} - (Y - \hat Y)\frac{\hat X}{\hat Y ^2} = \frac{\hat X}{\hat Y} + \frac{1}{\hat Y}(X - \frac{\hat X}{\hat Y}Y)$$

Исходя из этого, мы получаем выборку для одной случайной величины: $T_i - const * C_i$, для нее считаем доверительный интервал с помощью ttest, домножаем на размер выборки, чтобы получить доверительный интервал на суммы и затем модифицируем его, разделив на $\hat Y$ и прибавив $\frac{\hat X}{\hat Y}$, получим искомый ДИ

----

```python
def simple_relative_ttest_CI(test, control, alpha=0.05):
    """
        Построение оносительного доверительного интервала в случае выборок одинакового ращзмера: ET/EC - 1
    """
    assert len(test) == len(control)
    left_bound = 0
    right_bound = 0

    # Ваш код
    c1 = sum(test) / sum(control)
    c2 = 1/sum(control)
    g = np.array(test) - c1 * np.array(control)
    l_b, r_b = sps.t.interval(confidence=1 - alpha, loc=np.mean(g), df=len(g)-1, scale=sps.sem(g))
    left_bound = l_b*len(g)*c2 + c1
    right_bound = r_b*len(g)*c2 + c1
    return left_bound, right_bound
```

А теперь проверьте раш критерий на Монте-Карло: лежит ли истинный тносительный эффект в доверительном интервале, или нет? Изменять код генерации выборок ниже не надо.

```python
np.random.seed(8)

bad_cnt = 0

mc_size= 1000
test_size = 1000
control_size = 1000
effect = 0.6

for _ in range(mc_size):
    test = sps.expon(loc=0, scale=50).rvs(test_size) * (1 + effect)
    control = sps.expon(loc=0, scale=50).rvs(control_size)

    left_bound, right_bound = simple_relative_ttest_CI(test, control, alpha=0.05)
    if not (left_bound < 1 + effect < right_bound):
        bad_cnt += 1

# ваш код далее
l_b, r_b = proportion_confint(count = bad_cnt, nobs = mc_size, alpha=0.05, method='wilson')
print(f"FPR: {round(bad_cnt/mc_size, 3)}, [{round(l_b, 3)}, {round(r_b, 3)}]")
```

FPR приемлемый

Решите задачу в общем случае: что, если выборки теста и контроля разного размера? Как здесь применить линеаризацию (решать нужно именно через нее), если нельзя однозначно сопоставить элемент в тесте и в контроле? -бакеты?

Распишите ваш критерий: как он будет работать? На каком трюке он будет основан?

Основная проблема - разные размеры выборок. Если мы бы устранили эту проблему, то применили бы `simple_relative_ttest_CI`, который реализован выше. 

Итак, как уравнять размеры выборок, при этом не наломав дров?

Я реализовал способ, который описан в статье в условии. Я разбиваю выборки по бакетам, количество бакетов я задал 200. Это делается случайно - я беру остаток от индекса, перед этим можно конечно сделать random.shuffle, но мы ниже рассматриваем случайные сэмплирование выборок, поэтому этого можно не делать. 

Вот мы посчитали среднее по бакетам для теста и для контроля, получили 2 выборки равного размера (размер - количество бакетов), первая описывает тест, вторая описывает контроль.

Мы решили проблему, далее применим `simple_relative_ttest_CI` и мы выиграли.

```python
def full_relative_ttest_CI(test, control, alpha=0.05):
    """
        Построение оносительного доверительного интервала в случае выборок разного ращзмера: ET/EC - 1
    """
    left_bound = 0
    right_bound = 0
    B = 200
    g_test = np.zeros(B) # бакеты для тестовой выборки
    g_test_sizes = np.zeros(B)
    g_control = np.zeros(B) # бакеты для контроля
    g_control_sizes = np.zeros(B)
    
    test_size = len(test)
    control_size = len(control)
    
    # в бакетах я храню сумму
    for i in range(test_size):
        g_test[i % B] += test[i]
        g_test_sizes[i % B] += 1
    
    for i in range(control_size):
        g_control[i % B] += control[i]
        g_control_sizes[i % B] += 1
     
    # сформированную сумму я делю на количество значений, попавших в бакет
    test = g_test / g_test_sizes
    control = g_control.T / g_control_sizes
    
    left_bound, right_bound = simple_relative_ttest_CI(test, control)
    
    return left_bound, right_bound
```

```python
np.random.seed(8)

bad_cnt = 0

mc_size= 1000
test_size = 20000
control_size = 1000
effect = 0.6

for _ in range(mc_size):
    test = sps.expon(loc=0, scale=50).rvs(test_size) * (1 + effect)
    control = sps.expon(loc=0, scale=50).rvs(control_size)

    left_bound, right_bound = full_relative_ttest_CI(test, control, alpha=0.05)
    if not (left_bound < 1 + effect < right_bound):
        bad_cnt += 1

# ваш код далее
l_b, r_b = proportion_confint(count = bad_cnt, nobs = mc_size, alpha=0.05, method='wilson')
print(f"FPR: {round(bad_cnt/mc_size, 3)}, [{round(l_b, 3)}, {round(r_b, 3)}]")
```

Отличный FPR!

Реализуем bootstrap

```python
def theta_func(test, control):
    return np.mean(test)/np.mean(control)

def get_boot_sample_CI(theta_func, test, control, B=10000, alpha=0.05):
    theta_estim_array = []
    for _ in range(B):
        test_gen = np.random.choice(test, replace=True, size=len(test))
        control_gen = np.random.choice(control, replace=True, size=len(control))
        theta_estim_array.append(theta_func(test_gen, control_gen))
    l_b, r_b = np.quantile(theta_estim_array, [alpha/2, 1-alpha/2])
    return l_b, r_b
```

```python
for _ in tqdm(range(mc_size)):
    test = sps.expon(loc=0, scale=50).rvs(test_size) * (1 + effect)
    control = sps.expon(loc=0, scale=50).rvs(control_size)

    left_bound, right_bound = get_boot_sample_CI(theta_func, test, control, alpha=0.05, B=1000)
    if not (left_bound < 1 + effect < right_bound):
        bad_cnt += 1

l_b, r_b = proportion_confint(count = bad_cnt, nobs = mc_size, alpha=0.05, method='wilson')
print(f"FPR: {round(bad_cnt/mc_size, 3)}, [{round(l_b, 3)}, {round(r_b, 3)}]")
```

### Пункт D. 3 балла

Но и это еще не все: мы рассмотрим с вами, как построить более мощный критерий, основываясь на данных с пред периода.

Поговорим о критерии [постнормировки](https://habr.com/ru/companies/avito/articles/571096/).

В данном случае у нас есть 4 выборки:

- `test_before, control_before` &mdash; значения вашей метрики от пользователя на пред. эсперементальном периоде. Выборки в тесте и в контроле будут однородны.
- `test, control` &mdash; значения метрики от пользователя после начала AB-теста. Выборки теперь уже из разных распределений!


Нужно построить критерий для 

$$
    H_0: \dfrac{ET}{ET_b} / \dfrac{EC}{EC_b} - 1 = 0,
$$

а точнее доверительный интервал для этой гипотезы

Сначала опишите ваш алгоритм

--------

Так как в условии задания нет ограничений на методы, я буду использовать бутсрап. Чтобы хоть как-то ускорить подсчет доверительного интервала я буду применять бутстрап, используя разделение по батчам. 

Наша `theta_func` в данном примере: $$ \frac{\overline T}{\overline T_b} / \frac{\overline C}{\overline C_b} - 1 $$

Также при формировании случайной выборки с возвращением, надо учитывать, что мы работаем с юзерами, поэтому надо сначала задать список юзеров (также случайно с возвращением), а потом уже по этим юзерам считать средние для `theta_func`

Когда мы получили выборку, отбираем ДИ на эффект с помощью перцентильного интервала

```python
def post_normed_ttest(control, test, control_before, test_before, alpha=0.05):
    """
        Построение критерия пост-нормировки.
    """
    left_bound = 0
    right_bound = 0
    theta_func = lambda C, T, C_b, T_b: (T/T_b) / (C/C_b) - 1
    
    stat_sample = []
    batch_sz = 100
    boot_samples_size = 1000
    
    for i in range(0, boot_samples_size, batch_sz):
        control_ind = np.arange(len(control))
        test_ind = np.arange(len(test))
        
        # выбираем индексы для теста и контроля
        control_ind_sample = np.random.choice(control_ind, size=(len(control), batch_sz), replace=True)
        test_ind_sample = np.random.choice(test_ind, size=(len(test), batch_sz), replace=True)
        
        # считаем средние для theta_func
        C = np.mean(control[control_ind_sample], axis=0)
        T = np.mean(test[test_ind_sample], axis=0)
        C_b = np.mean(control_before[control_ind_sample], axis=0)
        T_b = np.mean(test_before[test_ind_sample], axis=0)

        stat_sample += list(theta_func(C, T, C_b, T_b))
    
    left_bound, right_bound = np.quantile(stat_sample, [0.025, 0.975])

    return left_bound, right_bound
```

Теперь давайте проверим ваш критерий:

```python
np.random.seed(8)

bad_cnt = 0

mc_size= 1000
test_size = 20000
control_size = 1000
effect = 0.3

for _ in tqdm(range(mc_size)):
    test_before = sps.expon(loc=0, scale=50).rvs(test_size)
    control_before = sps.expon(loc=0, scale=50).rvs(control_size)

    test = (test_before + sps.norm(loc=0, scale=10).rvs(test_size)) * (1 + effect)
    control = (control_before + sps.norm(loc=0, scale=20).rvs(control_size))

    left_bound, right_bound = post_normed_ttest(control, test, control_before, test_before, alpha=0.05)
    if not (left_bound < effect < right_bound):
        bad_cnt += 1

l_b, r_b = proportion_confint(count = bad_cnt, nobs = mc_size, alpha=0.05, method='wilson')
print(f"FPR: {round(bad_cnt/mc_size, 3)}, [{round(l_b, 3)}, {round(r_b, 3)}]")
```

Критерий работает отлично

### Пункт E. 2 балла

Как вы думаете, зачем нужна постнормировка на практике? Ответ поясните на практическом примере.

Разберем пример, мы разделили тест и контроль случайным образом. Получилось так, что тест и контроль получились несбалансированными - у теста например среднее в 2 раза больше, чем у контроля, но мы этого не знаем, так как мы решили не учитывать прошлое. Затем применили обычный t-test и получили прокрас. Но на самом деле ничего не изменилось, а прокрас получили просто из-за того, что среднее на тесте в 2 раза больше и так. 

Для этого и нужна постнормировка. Чтобы балансировать выборки для критерия.

Симмитируем пример из жизни некоректного (для обычного ttest, но не для постнормировки) разбиения на тест и контроль:


```python
np.random.seed(8)

bad_cnt_ttest = 0
bad_cnt_postn = 0


mc_size= 1000
test_size = 100
control_size = 5000
# Для наглядности добавлю нулевой эффект
effect = 0

for _ in tqdm(range(mc_size)):
    
    # Предположим, что test и control так распределены, у контроля матожидание больше на 40% по сравнению с тестом
    test_before = sps.expon(loc=0, scale=50).rvs(test_size)
    control_before = sps.expon(loc=20, scale=50).rvs(control_size)

    # Иммитируем данные эксперимента
    test = (test_before + sps.norm(loc=0, scale=10).rvs(test_size)) * (1 + effect)
    control = (control_before + sps.norm(loc=0, scale=20).rvs(control_size))

    # Для критерия постнормировки
    left_bound_post, right_bound_post = post_normed_ttest(control, test, control_before, test_before, alpha=0.05)
    if not (left_bound_post < effect < right_bound_post):
        bad_cnt_postn += 1
    
    # Для ttest
    pvalue_ttest = sps.ttest_ind(test, control).pvalue
    if pvalue_ttest < 0.05:
        bad_cnt_ttest += 1

l_b, r_b = proportion_confint(count = bad_cnt_ttest, nobs = mc_size, alpha=0.05, method='wilson')
print(f"T-test: {round(bad_cnt_ttest/mc_size, 3)}, [{round(l_b, 3)}, {round(r_b, 3)}]")

l_b, r_b = proportion_confint(count = bad_cnt_postn, nobs = mc_size, alpha=0.05, method='wilson')
print(f"Постнормировка: {round(bad_cnt_postn/mc_size, 3)}, [{round(l_b, 3)}, {round(r_b, 3)}]")
```

Получаем, что ttest не корректен вообще в данном случае, мы не добавили никакого `effect`, а t-test ошибается в том, что мы его добавили в 96% случаях, что очевидно, так как тест и контроль несбалансированные.

А вот как раз критерий постнормировки это учитывает, поэтому ошибка приемлемая, критерий корректен.
